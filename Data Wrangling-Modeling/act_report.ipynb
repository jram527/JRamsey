## Report: act_report

This report outlines the insights and visualizations of the archive from the Twitter account WeRateDogs. The analysis began interacting with three datasets: The main Twitter archive, the Tweet Image Predictions, and the Twitter JSON. The Twitter archive is a comma-separated values file containing approximately 5000 tweets. The Twitter API/JSON is a scrape of retweet and favorite count data downloaded programmatically, and the Image Predictions file is a tab-separated values file containing tweet data from a neural network. Each of the datasets underwent the process of gathering, accessing, cleaning, and storing using the pandas, requests, os, json, NumPy, seaborn, and matplotlib libraries in Jupytr Notebooks. 

#### 1. Out of the top ten dog names, the most popular name is "Charlie".

Using the newly wrangled data, the “name” column was used to see the frequencies of all dog names. This was used to determine the top ten most popular dog names in this dataset. This insight was generated using a bar chart. “Charlie” was the most frequent name that was tweeted.


#### 2. “Duddles” was the most favored. 

By grouping by "name" and "favorite_count", “Duddles” was found to be the highest name among the names with the most favorited tweets. It implies that although “Charlie” has most associated tweets, “Duddles” had the most interactions through favorites.


#### 3. The Golden Retriever had the most accurate image predictions.

The column p1_dog represents the initial prediction of the dog breed by the neural network. “True” indicates the prediction was correct, and “False” indicates an incorrect prediction. By querying the p1_dog column for “True”, the associated breeds with the correct prediction can be plotted.


#### Visualization of Dog Stage Tweets

Lastly, the Dog Stage barplot was constructed using the "dog_stage" column. This visualizes that most users associated with pupper.
